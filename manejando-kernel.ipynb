```python
import numpy as np
import matplotlib.pyplot as plt 
import scipy.ndimage as nd 
from skimage import  io, color 
```

```python
img = io.imread('../input/cnn-data-sources/yo.png')
img_gray = color.rgb2gray(img)
```

```python
plt.imshow(img_gray)
```

## Defincin de los kernel

```python
kernel_vertical = np.array([[-1,0,1],
                  [-1,0,1],
                   [-1,0,1]
                  ])
```

```python
kernel_horizontal = kernel_vertical.T
```

```python
kernel_neg = np.array([[0,0,0],
                      [0,-1,0],
                       [0,0,0]
                      ])
```

```python
kernel_sobel = np.array([[-1, -2, -1],
                      [0,0,0],
                        [1, 2,1]
                       ])
```

## Aplicacin de cada kernel

```python
img_bordes_ver = nd.convolve(img_gray, kernel_vertical)
```

```python
img_bordes_hor = nd.convolve(img_gray, kernel_horizontal)
```

```python
img_neg = nd.convolve(img_gray, kernel_neg)
```

```python
img_sobel = nd.convolve(img_gray, kernel_sobel )
```

```python
fig, axes = plt.subplots(1,2, figsize = (15,10))
axes[0].imshow(img_gray, cmap=plt.cm.gray)
axes[0].set_title('Original')
axes[0].axis('off')

axes[1].imshow(img_bordes_ver, cmap=plt.cm.gray)
axes[1].set_title('Convolution')
axes[1].axis('off')
```

```python
fig, axes = plt.subplots(1,2, figsize = (15,10))
axes[0].imshow(img_gray, cmap=plt.cm.gray)
axes[0].set_title('Original')
axes[0].axis('off')

axes[1].imshow(img_bordes_hor, cmap=plt.cm.gray)
axes[1].set_title('Convolution')
axes[1].axis('off')
```

```python
fig, axes = plt.subplots(1,2, figsize = (15,10))
axes[0].imshow(img_gray, cmap=plt.cm.gray)
axes[0].set_title('Original')
axes[0].axis('off')

axes[1].imshow(img_neg, cmap=plt.cm.gray)
axes[1].set_title('Convolution')
axes[1].axis('off')
```
